Às vezes, você precisa se comunicar entre os threads . Para fazer isso, você pode usar um bloqueio de exclusão mútua (mutex) e uma variável booleana.

```python
from threading import Thread, Event
from time import sleep

def task(event: Event, id: int) -> None:
    print(f'Thread {id} iniciada. Esperando pelo sinal....')
    event.wait()
    print(f'Sinal recebido. A thread {id} esta completa.')

def main() -> None:
    event = Event()

    t1 = Thread(target=task, args=(event,1))
    t2 = Thread(target=task, args=(event,2))

    t1.start()
    t2.start()

    print('Bloqueando a thread principal por 3 segundos...')
    sleep(3) 
    event.set()

if __name__ == '__main__':
    main()
```
